```python
# Let's examine the case of NaN. There are many entries like this.
# We will reclassify the values of NaN based on columns 'Injury' and 'Species' since it seems the data entrier
# had put some keywords such as FATAL and Death on these columns and could lead us to improve our dataset.

# Let's define a function to apply to our dataset, where we classify the values in Fatal_adjusted based on
# columns 'Injury' and 'Species', when they do have the info 'death' or 'fatal'.

shark_attacks['Fatal_adjusted'] = 'Test1'
print(shark_attacks['Fatal_adjusted'])

fatal_list = re.findall('[fF][aA][tT][aA][lL]', shark_attacks['Injury'])
print(fatal_list)

def classifier(x):
    
    fatal_list = re.findall('[fF][aA][tT][aA][lL]', x)
    if fatal_list in shark_attacks['Injury'] or shark_attacks['Species']:
        return 'Y'
    
    elif re.findall('[dD][eE][aA][tT][hH]', x) in shark_attacks['Injury'] or shark_attacks['Species']:
        return 'Y'
    
    else:
       pass


# Let's create another column to adjust the data from 'Fatal (Y/N)', called 'Fatal_adjusted'.
    
shark_attacks['Fatal_adjusted'] = shark_attacks['Injury'].apply(classifier)

print(shark_attacks['Fatal_adjusted'])

```
